### Exercise 6: Histogram Customization

```python
# TASK: Create a histogram for the following data: data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5].
# Customize the histogram with a title, labels for the x-axis, and a specific color for the bars.
```

```python
import matplotlib.pyplot as plt
import numpy as np

try:
    # Data for plots
    x = np.arange(1, 6)
    y = x * 2
    categories = ['A', 'B', 'C', 'D']
    values = [5, 7, 3, 9]
    labels = ['Python', 'Java', 'C++', 'JavaScript']
    sizes = [40, 25, 20, 15]
    data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]

    # Create subplots
    fig, axes = plt.subplots(2, 3, figsize=(15, 10))

    # Line plot
    axes[0, 0].plot(x, y, marker='o', linestyle='-', color='b')
    axes[0, 0].set_title("Line Plot")
    axes[0, 0].set_xlabel("X Values")
    axes[0, 0].set_ylabel("Y Values")
    axes[0, 0].grid(True)

    # Bar chart
    axes[0, 1].bar(categories, values, color=['r', 'g', 'b', 'y'])
    axes[0, 1].set_title("Bar Chart")
    axes[0, 1].set_xlabel("Categories")
    axes[0, 1].set_ylabel("Values")

    # Scatter plot
    axes[0, 2].scatter(x, y, color='b')
    for i, j in zip(x, y):
        axes[0, 2].annotate(f'({i}, {j})', (i, j), textcoords="offset points", xytext=(5,5), ha='right')
    axes[0, 2].set_title("Scatter Plot")
    axes[0, 2].set_xlabel("X Values")
    axes[0, 2].set_ylabel("Y Values")

    # Pie chart
    axes[1, 0].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['blue', 'red', 'green', 'orange'])
    axes[1, 0].set_title("Pie Chart")

    # Histogram
    axes[1, 1].hist(data, bins=5, color='purple', edgecolor='black')
    axes[1, 1].set_title("Histogram")
    axes[1, 1].set_xlabel("Data Values")
    axes[1, 1].set_ylabel("Frequency")

    # Adjust layout
    plt.tight_layout()

    # Show the plots
    plt.show()

except Exception as e:
    print(f"An error occurred: {e}")
```
